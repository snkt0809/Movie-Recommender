```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head()
```

```python

movies.dropna(inplace = True)
movies.isnull().sum()
```

```python
movies =movies.drop_duplicates()
movies
```

```python

movies.duplicated().sum()
```

```python
import ast 
def process(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

```

```python
movies['keywords'] = movies['keywords'].apply(process)
movies['keywords']
```

```python
movies['genres'] = movies['genres'].apply(process)
```

```python
movies['genres']
```

```python
def givetop3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(str(obj)):
        if counter !=3:
            L.append(i['name'])
            counter += 1 
        else : 
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(givetop3)
print(movies['cast'])
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(str(obj)):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
movies['crew']
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['overview']
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['genres']
```

```python
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies['cast']
```

```python
movies['keywords']
```

```python
movies['crew']
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id', 'title', 'tags']]
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : " ".join(x))
new_movies['tags']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    lis = []
    for i in text.split(): 
        lis.append(ps.stem(i))
    return " ".join(lis)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
new_movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity.shape[0]
ranked_table = []
for i in range(len(similarity)):
    appendable = sorted(list(enumerate(similarity[i])), reverse = True, key = lambda x:x[1])[1:11]
    ranked_table.append(appendable)
ranked_table
```

```python
ranked_table[0][1:7]
```

```python
import pickle
pickle.dump(ranked_table, open("ranked_table.pkl", "wb"))
```

```python
def recommend(movie):
    selected_movie_id = new_movies[new_movies['title'] == movie].index[0]
    similarities = similarity[selected_movie_id]
    movies_list= sorted(list(enumerate(similarities)), reverse = True, key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
    
    
```

```python
def recommend(movie):
    selected_movie_id = new_movies[new_movies['title'] == movie].index[0]
    movies_list =ranked_table[selected_movie_id][0:5]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```
